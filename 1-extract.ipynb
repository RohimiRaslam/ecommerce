```python
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd
from bs4 import BeautifulSoup
```

```python
base_url = "https://www.amazon.co.jp"

options = webdriver.FirefoxOptions()
options.add_argument('--headless')
driver = webdriver.Firefox(options=options)
search_url = 'https://www.amazon.co.jp/-/en/ranking?type=top-sellers&ref_=nav_cs_bestsellers'
driver.get(search_url)

html_source = driver.page_source
soup = BeautifulSoup(html_source, 'html.parser')

categories = soup.find_all("div" , class_="_p13n-zg-nav-tree-all_style_zg-browse-item__1rdKf _p13n-zg-nav-tree-all_style_zg-browse-height-small__nleKL")
headers = [header.text for header in categories]
categories_urls = [base_url + category.find('a').get('href') for category in categories]
```

# Amazon_Devices


```python
Amazon_Devices_Accessories_url = categories_urls[0]
driver.get(Amazon_Devices_Accessories_url)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
driver.implicitly_wait(30)
ama_dev_html = driver.page_source
ama_dev_soup = BeautifulSoup(ama_dev_html, 'html.parser')
ama_dev_products = ama_dev_soup.find_all("div" , class_="p13n-sc-uncoverable-faceout")
ama_dev_rank = ama_dev_soup.find_all('div' , class_='a-section zg-bdg-body zg-bdg-clr-body aok-float-left')
```

```python
product_dict = {'name' : [] , 'rank' : [] ,  'rating_(out_of_5_stars)' : [] ,'no_of_votes' : [] , 'price' : [] , 'link' : [] , 'available_(Yes/No)' : []}

for i in range(len(ama_dev_products)):
    rank = ama_dev_rank[i].find('span' , class_='zg-bdg-text')
    name_raw = ama_dev_products[i].find('div' , class_='_cDEzb_p13n-sc-css-line-clamp-3_g3dy1')
    name = name_raw.text if name_raw else ""
    rating = ama_dev_products[i].find('span' , class_='a-icon-alt')
    price_raw = ama_dev_products[i].find('span' , class_='p13n-sc-price')
    price = price_raw.text if price_raw else "N/A"
    rating_vote_raw = ama_dev_products[i].find('span' , class_='a-size-small')
    rating_vote = rating_vote_raw.text
    link = base_url + ama_dev_products[i].find('a' , class_='a-link-normal aok-block').get('href')
        

    product_dict['name'].append(name.strip())
    product_dict['rank'].append(rank.text.strip())
    product_dict['rating_(out_of_5_stars)'].append(rating.text)
    product_dict['price'].append(price)
    product_dict['no_of_votes'].append(rating_vote.strip())
    product_dict['link'].append(link)  
    product_dict['available_(Yes/No)'].append("Yes") if price_raw else product_dict['available_(Yes/No)'].append("No")

df_ama_dev = pd.DataFrame(product_dict)
```
